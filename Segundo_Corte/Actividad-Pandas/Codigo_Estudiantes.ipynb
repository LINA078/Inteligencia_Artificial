##CODIGO COMPLETO NOTAS DE ESTUDIANTES


```python
import pandas as pd
#crear datos de estudiantes
Datos={"Estudiantes": ["Ana", "Lucas", "Carlos", "Maria", "Lucia"], "Nota_1": [60, 56, 48, 74, 65],
       "Nota_2":[100, 70,65,90,60],
       "Nota_3":[60,85,54,55,98]

       }
       #crear DataFrame
df = pd.DataFrame(Datos)
#calcular el promedio de cada estudiante
df["Promedio"]= (df["Nota_1"] +  df["Nota_2"] + df["Nota_3"])/3

#determinar si aprueba o no aprueba
df["Resultado"] = df["Promedio"].apply(
    lambda x: "Aprueba" if x >= 60 else "No aprueba"
)

#calcular el promedio general del grupo


Promedio_General = df["Promedio"].mean()

Estudiante_Mayor = df.loc[df["Promedio"].idxmax(), "Estudiantes"]
Nota_Mayor = df["Promedio"].max()

Estudiante_Menor = df.loc[df["Promedio"].idxmin(), "Estudiantes"]
Nota_Menor = df["Promedio"].min()

print("El promedio general del grupo es:", round(Promedio_General, 2))
print("El estudiante con mayor promedio es:", Estudiante_Mayor, "con una nota de:", round(Nota_Mayor, 2))
print("El estudiante con menor promedio es:", Estudiante_Menor, "con una nota de:", round(Nota_Menor, 2))

#guardar archivo CSV
df.to_csv("Notas_Estudiantes.csv" ,
index=False)
 #leer archivo CSV
Archivo_Notas = pd.read_csv("Notas_Estudiantes.csv")
#mostrar tabla final
Archivo_Notas
```
