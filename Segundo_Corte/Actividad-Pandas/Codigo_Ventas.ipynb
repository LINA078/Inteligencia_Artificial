##CODIGO COMPLETO DE VENTAS


```python
import pandas as pd

# Crear datos sintéticos de productos
Datos_Sinteticos = {
    "Producto": ["Camisa", "Pantalon", "Zapatos", "Bolso", "Gorra"],
    "Precio": [35000, 60000, 40000, 12000, 55000],
    "Cantidad_Vendida": [20, 8, 18, 6, 45]
}

# Crear DataFrame
Ventas = pd.DataFrame(Datos_Sinteticos)

# Calcular el total vendido por cada producto
Ventas["Total_Vendido"] = Ventas["Cantidad_Vendida"] * Ventas["Precio"]

# Calcular el total general de ventas
Total_General = Ventas["Total_Vendido"].sum()

# Identificar el producto más vendido
Producto_Mas_Vendido = Ventas.loc[Ventas["Cantidad_Vendida"].idxmax(), "Producto"]
Cantidad_Mayor = Ventas["Cantidad_Vendida"].max()

# Identificar el producto menos vendido
Producto_Menos_Vendido = Ventas.loc[Ventas["Cantidad_Vendida"].idxmin(), "Producto"]
Cantidad_Menor = Ventas["Cantidad_Vendida"].min()

# Mostrar resultados
print("El total general de ventas es:", Total_General)
print("El producto más vendido es:", Producto_Mas_Vendido, "con", Cantidad_Mayor, "unidades vendidas")
print("El producto menos vendido es:", Producto_Menos_Vendido, "con", Cantidad_Menor, "unidades vendidas")

# Guardar archivo CSV
Ventas.to_csv("Ventas.csv", index=False)

print("El archivo CSV fue creado correctamente")

# Leer archivo CSV
Archivo_Ventas = pd.read_csv("Ventas.csv")

# Mostrar tabla final
Archivo_Ventas
```
